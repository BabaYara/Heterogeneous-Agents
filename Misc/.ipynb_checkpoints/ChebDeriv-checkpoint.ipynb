```julia
using BasisMatrices
using LaTeXStrings
using Plots; pyplot();

#basis = Basis(SplineParams(linspace(0.1, 20, 10), 0, 3))
basis = Basis(ChebParams(12,-1,1))
S, (A, ) = nodes(basis)
Φ = BasisMatrix(basis, Expanded()).vals[1]

V = [1.09661, 4.00659, 4.62933, 4.83924, 5.02284, 5.19641, 5.35062, 5.48095, 5.58552, 5.6638, 5.71583, 5.74177]
c = Φ \ V;
```

```julia
plot(A, V, label = L"$V$", alpha=0.3)
plot!(A, BasisMatrix(basis, Expanded(), S).vals[1] * c, label = L"$\hat{V}$")
scatter!(A, V)
```

```julia
plot(A, BasisMatrix(basis, Expanded(), S, [1]).vals[1] * c, label = L"$\hat{V}\,'$")
scatter!(A, zeros(A))
```
